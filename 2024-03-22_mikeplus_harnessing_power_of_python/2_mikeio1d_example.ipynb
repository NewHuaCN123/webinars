# Example 2 - Read .res1d, .xns11, .res, .resx, and .out files.

#### Res1d files (and .res, .resx, .out)

```python
# Imports

from mikeio1d import Res1D
```

```python
# Open file

res = Res1D('./data/blue_beach_network.res1d')
res
```

```python
# Print meta info

res.info()
```

```python
# Read everything.

res.read()
```

```python
# Read results for reaches, in several different ways.

# res.reaches
# res.reaches['Link']
# res.reaches['Link_16'].Discharge.read()
# res.reaches['Link_16'].read()
# res.reaches['Link_16'].plot()
# res.reaches.read()
# res.reaches.Discharge.read()
res.reaches.Discharge.read(column_mode='compact')
```

```python
# Same applies for nodes, as well as for catchments.

res.nodes
res.nodes.read(column_mode="compact").describe().T
```

```python
# Plot the network

res.network.to_geopandas().plot()
```

```python
# Make a simple plot of the maximum discharge.

gdf = res.reaches.to_geopandas("max")
gdf.plot("max_Discharge", legend=True)
```

```python
# Plot catchment runoff

res_catch = Res1D("data/blue_beach_runoff.res1d")
res_catch.catchments.to_geopandas(agg="max").plot("max_TotalRunOff", legend=True)
```

```python
# Export to shapefile, then go use it in QGIS or something.

gdf.to_file("max_discharge_reaches.shp")
```

```python
# Or explore directly within your notebook

gdf.explore(tiles="cartodb positron", column="max_Discharge", vmax=0.5)
```
